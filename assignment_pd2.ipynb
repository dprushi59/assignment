```python
#1
# Here are five commonly used functions in the pandas library with their execution examples:

# 1. read_csv(): This function is used to read data from a CSV file and create a DataFrame.
import pandas as pd

# # Read CSV file and create a DataFrame
data = pd.read_csv('data2.csv')
print(data.head())

# 2. head(): This function is used to display the first few rows of a DataFrame.
import pandas as pd

# Create a DataFrame
data = pd.DataFrame({'Name': ['John', 'Jane', 'Mike', 'Sarah'],'Age': [25, 28, 32, 30]})

# # Display the first few rows
print(data.head())

# 3. info(): This function provides a summary of a DataFrame, including the data types, column names, and non-null counts.
import pandas as pd

# # Create a DataFrame
data = pd.DataFrame({'Name': ['John', 'Jane', 'Mike', 'Sarah'],
                     'Age': [25, 28, 32, 30]})

# # Display summary information
data.info()

#4. groupby(): This function is used for grouping data based on one or more columns of a DataFrame.
import pandas as pd

# # Create a DataFrame
data = pd.DataFrame({'Name': ['John', 'Jane', 'Mike', 'Sarah'],
                    'Age': [25, 28, 32, 30],
                     'Department': ['Sales', 'Marketing', 'Sales', 'Marketing']})

# # Group data by Department and calculate average age
grouped_data = data.groupby('Department')['Age'].mean()
print(grouped_data)

#5. merge(): This function is used to merge two or more DataFrames based on a common key or column.
import pandas as pd

# # Create two DataFrames
df1 = pd.DataFrame({'Name': ['John', 'Jane', 'Mike', 'Sarah'],
                     'Age': [25, 28, 32, 30],
                    'Department': ['Sales', 'Marketing', 'Sales', 'Marketing']})

df2 = pd.DataFrame({'Name': ['John', 'Jane', 'Mike', 'Sarah'],
                    'Salary': [5000, 6000, 5500, 7000]})

# # Merge the DataFrames based on 'Name' column
merged_data = pd.merge(df1, df2, on='Name')
print(merged_data)

# Please note that the above examples assume you have the required data or CSV file in the specified format. You may need to adjust the code according
#to your specific use case.
```

```python
#5
# In the context of pandas DataFrame, the functions `DataFrame.size()` and `DataFrame.shape()` serve different purposes:

# 1. DataFrame.size():
#    - `DataFrame.size` returns the total number of elements in the DataFrame, which is equal to the product of the number of rows and columns.
#    - It represents the total size or length of the DataFrame.
#    - The return value is an integer.
#    - It is calculated as `DataFrame.size = DataFrame.shape[0] * DataFrame.shape[1]`.

# 2. DataFrame.shape():
#    - `DataFrame.shape` returns a tuple containing the dimensions of the DataFrame, i.e., the number of rows and columns.
#    - It provides information about the structure of the DataFrame.
#    - The return value is a tuple `(rows, columns)`.
#    - The number of rows is represented by `DataFrame.shape[0]`, and the number of columns is represented by `DataFrame.shape[1]`.
```

```python
#9
def stats(df,col):
    mean = df[col].mean()
    median = df[col].median()
    std = df[col].std()
    d = {'mean' : mean , 'median': median , 'std' : std}
    return d
```

```python
stats(df1,'Values')
```

```python
def weekday(df,col,new_col):
    l = ['Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday']
    l1 = l
    for i in range(5):
        l = l + l1
    df[new_col] = pd.Series(l)
    return df
```
